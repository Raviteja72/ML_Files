```python
img1 = cv2.imread('photo-1441260038675-7329ab4cc264.webp')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('photo-1548438294-1ad5d5f4f063.jpeg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img1 = cv2.resize(img1,(600,600))
plt.imshow(img1)
```

```python
LARGE_IMG = img1
SMALL_IMG = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + SMALL_IMG.shape[1]
y_end = y_offset + LARGE_IMG.shape[0]
```

```python
x_end
```

```python
LARGE_IMG[y_offset:y_end,x_offset:x_end] = SMALL_IMG
plt.imshow(LARGE_IMG)
```

```python
img2 = cv2.imread('IMG_20190329_220029.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img2,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2)
plt.imshow(mask_inv)
```

```python
fg = cv2.bitwise_or(img2,img2,mask=mask_inv)
plt.imshow(fg)
```
