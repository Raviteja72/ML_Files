```python
img = cv2.imread('jack_dorsey.jpg')

font = cv2.FONT_HERSHEY_DUPLEX

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(img,text='JACK DORSEY',org=(5,155),fontFace=font,fontScale=4,color=(132,255,255),thickness=4,lineType=cv2.LINE_AA)
plt.figure(figsize=(12,8))
plt.imshow(img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
verices = np.array([ [300,400], [400,300], [300,200], [200,300] ],dtype=np.int32)
```

```python
verices
```

```python
verices.shape
```

```python
vertices = verices.reshape((-1,1,2))
print(verices)
```

```python
vertices.shape
```

```python
cv2.polylines(blank_img,pts=[vertices],isClosed=True,color=(150,150,250),thickness=5)
cv2.line(blank_img,pt1=(300,200),pt2=(400,300),color=(132,255,255),thickness=4,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
